```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
```

```python
class ColorFilter():
    @staticmethod
    def invert(array):
        arr = np.copy(array)
        arr[:,:,0:3] = 1.0 - array[:,:,0:3] 
        return(arr)
    
    @staticmethod
    def to_blue(array):
        arr = np.copy(array)
        shp = array.shape
        print(shp)
        arr[:,:,0] = np.zeros((shp[0], shp[1]))
        arr[:,:,1] = np.zeros((shp[0], shp[1]))
        return(arr)
    
    @staticmethod
    def to_red(array):
        arr = np.copy(array)
        shp = array.shape
        print(shp)
        arr[:,:,1] = np.zeros((shp[0], shp[1]))
        arr[:,:,2] = np.zeros((shp[0], shp[1]))
        return(arr)
    
    @staticmethod
    def to_green(array):
        arr = np.copy(array)
        shp = array.shape
        print(shp)
        arr[:,:,0] = np.zeros((shp[0], shp[1]))
        arr[:,:,2] = np.zeros((shp[0], shp[1]))
        return(arr)
    
    @staticmethod
    def to_celluloid(arr):
        a = np.copy(arr)
        lsp_R = np.linspace(arr[:,:,0].min(), arr[:,:,0].max(), 5)
        lsp_G = np.linspace(arr[:,:,1].min(), arr[:,:,1].max(), 5)
        lsp_B = np.linspace(arr[:,:,2].min(), arr[:,:,2].max(), 5)
        start_R = 0.5 * (lsp_R[0] + lsp_R[1])
        start_G = 0.5 * (lsp_G[0] + lsp_G[1])
        start_B = 0.5 * (lsp_B[0] + lsp_B[1])
        end_R = 0.5 * (lsp_R[-2] + lsp_R[-1])
        end_G = 0.5 * (lsp_G[-2] + lsp_G[-1])
        end_B = 0.5 * (lsp_B[-2] + lsp_B[-1])
        mid_R = np.linspace(start_R, end_R, 4)
        mid_G = np.linspace(start_G, end_G, 4)
        mid_B = np.linspace(start_B, end_B, 4)
        mask_R1 = (a[:,:,0] > lsp_R[0]) & (a[:,:,0] <= lsp_R[1])
        mask_R2 = (a[:,:,0] > lsp_R[1]) & (a[:,:,0] <= lsp_R[2])
        mask_R3 = (a[:,:,0] > lsp_R[2]) & (a[:,:,0] <= lsp_R[3])
        mask_R4 = (a[:,:,0] > lsp_R[3]) & (a[:,:,0] <= lsp_R[4])
        
        mask_G1 = (a[:,:,1] > lsp_G[0]) & (a[:,:,1] <= lsp_G[1])
        mask_G2 = (a[:,:,1] > lsp_G[1]) & (a[:,:,1] <= lsp_G[2])
        mask_G3 = (a[:,:,1] > lsp_G[2]) & (a[:,:,1] <= lsp_G[3])
        mask_G4 = (a[:,:,1] > lsp_G[3]) & (a[:,:,1] <= lsp_G[4])
        
        mask_B1 = (a[:,:,2] > lsp_B[0]) & (a[:,:,2] <= lsp_B[1])
        mask_B2 = (a[:,:,2] > lsp_B[1]) & (a[:,:,2] <= lsp_B[2])
        mask_B3 = (a[:,:,2] > lsp_B[2]) & (a[:,:,2] <= lsp_B[3])
        mask_B4 = (a[:,:,2] > lsp_B[3]) & (a[:,:,2] <= lsp_B[4])
        
        a[:,:,0][mask_R1] = mid_R[0]
        a[:,:,0][mask_R2] = mid_R[1]
        a[:,:,0][mask_R3] = mid_R[2]
        a[:,:,0][mask_R4] = mid_R[3]
        
        a[:,:,1][mask_G1] = mid_G[0]
        a[:,:,1][mask_G2] = mid_G[1]
        a[:,:,1][mask_G3] = mid_G[2]
        a[:,:,1][mask_G4] = mid_G[3]
        
        a[:,:,2][mask_B1] = mid_B[0]
        a[:,:,2][mask_B2] = mid_B[1]
        a[:,:,2][mask_B3] = mid_B[2]
        a[:,:,2][mask_B4] = mid_B[3]
        
        return(a)
    
    @staticmethod
    def to_grayscale(array):
        pass
```
